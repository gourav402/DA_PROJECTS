# Exploratory Data Analysis

```python
df.columns
```

## GENDER

```python
# BUYERS BASED ON GENDER
ax = sns.countplot(x= 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gen_sales = df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by= 'Amount',ascending = False)
gen_sales
```

```python
# PURCHASING POWER BASED ON GENDER
ax= sns.barplot(x = 'Gender',y='Amount',data = gen_sales )
```

### From the above graph we can see that most of the buyers are females and even the purchasing power of females are greater than men

## AGE

```python
df.columns
```

```python
# BUYERS BASED ON AGE GROUP
sns.countplot(data= df,x='Age Group')
```

```python
# BUYERS BASED ON AGE GROUP AND GENDER
ax= sns.countplot(data= df,x='Age Group',hue= 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# PURCHASING POWER ACCORDING TO AGE GROUP
age_sales = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)

sns.barplot(x='Age Group',y='Amount',data= age_sales)
```

### From the above graph we can see that most of the buyers are from age group 26-35 females

## MARITAL STATUS

```python
status_sales= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
status_sales
```

```python
# PURCHASING POWER ACCORDING TO MARITAL STATUS AND GENDER
sns.barplot(x= 'Marital_Status',y='Amount',hue='Gender',data=status_sales)
```

```python
# BUYERS BASED ON MARITAL STATUS
ax= sns.countplot(df,x='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

### From the above graphs we can see that most of the buyers are married woman and they have high purchasing power

## STATE

```python
# PURCHASING POWER OF TOP 10 STATES
state_sales= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data= state_sales,x='State',y='Amount')
```

```python
# TOP 10 STATES WITH HIGHEST ORDERS 
ax= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by= 'Orders',ascending=False).head(10)
aa= sns.barplot(data=ax,x='State',y='Orders')
for bars in aa.containers:
    aa.bar_label(bars)
    
```

```python
# TOP 10 STATES WITH lOWEST ORDERS 
ax= df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by= 'Orders',ascending=False).tail(10)
aa= sns.barplot(data=ax,x='State',y='Orders')
for bars in aa.containers:
    aa.bar_label(bars)
    
```

```python
# 10 LOWEST STATES WITH PURCHASING POWER
state_sales= df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= True).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data= state_sales,x='State',y='Amount')
```

### From the above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra, Karnataka.

# ZONE

```python
# ZONES ACCORDING TO PURCHASING POWER
zone_sales= df.groupby(['Zone'],as_index=False)['Amount'].sum().sort_values(by= 'Amount',ascending= False)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(data=zone_sales,x='Zone',y='Amount')
```

```python
# BUYERS ACCORDING TO ZONES
sns.countplot(data=df,x='Zone')
```

## OCCUPATION

```python
# BUYERS FROM DIFFERENT OCCUPATION
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# PURCHASING POWER OF BUYERS FROM DIFFERENT OCCUPATION
occ_sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=occ_sales,x='Occupation',y='Amount')
```

### From the above graphs we can see that most are the buyers from IT, Aviation and Healtha Care sectors

## PRODUCT CATEGORY

```python
# BUYERS ACCORDING TO PRODUCT CATEGORY
sns.set(rc={'figure.figsize':(24,7)})
ax= sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# CATEGORY WITH MOST SOLD PRODUCTS
prod_sales= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(19,7)})
sns.barplot(data=prod_sales,x='Product_Category',y='Amount')
```

### From the above graphs we can say that most of te sold product are from Food, Clothing and Elecronics category

# TOP 10 PERFORMING PRODUCTS

```python
top_prod= df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=top_prod,x='Product_ID',y='Orders')
```
